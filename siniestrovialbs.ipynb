```python
# Calcular el número total de víctimas
total_victimas = len(df)

# Filtrar el DataFrame para obtener solo las víctimas femeninas
victimas_femeninas = df[df['sexo'] == 'Femenino']

# Contar el número de víctimas femeninas
num_victimas_femeninas = len(victimas_femeninas)

# Filtrar el DataFrame para obtener solo las víctimas masculinas
victimas_masculinas = df[df['sexo'] == 'Masculino']

# Contar el número de víctimas masculinas
num_victimas_masculinas = len(victimas_masculinas)

# Calcular los porcentajes
porcentaje_femenino = (num_victimas_femeninas / total_victimas) * 100
porcentaje_masculino = (num_victimas_masculinas / total_victimas) * 100

# Imprimir los resultados
print("Número de víctimas femeninas:", num_victimas_femeninas)
print("Número de víctimas masculinas:", num_victimas_masculinas)
print("Porcentaje de víctimas femeninas:", porcentaje_femenino)
print("Porcentaje de víctimas masculinas:", porcentaje_masculino)
```

```python
# Contar el número de víctimas por municipio
victimas_por_municipio = df['municipio_nombre'].value_counts(dropna=False)

# Contar el número de casos sin especificar
num_sin_especificar = df['municipio_nombre'].isna().sum()

# Imprimir el resultado
print("Víctimas por municipio:")
print(victimas_por_municipio)


```

```python
# Calcular el porcentaje de casos sin especificar
porcentaje_sin_especificar = (num_sin_especificar / len(df)) * 100

# Calcular el porcentaje de casos por municipio
porcentaje_por_municipio = (victimas_por_municipio / len(df)) * 100

# Imprimir los resultados
print("Porcentaje de casos sin especificar:", porcentaje_sin_especificar)
print("\nPorcentaje de casos por municipio:")
print(porcentaje_por_municipio)
```

```python


# Ordenar los datos en orden descendente
conteo_por_tipo_via_sorted = conteo_por_tipo_via.sort_values(ascending=False)

# Crear una paleta de colores personalizada con tonos de rosa
custom_palette = sns.color_palette("pink", len(conteo_por_tipo_via_sorted))

# Crear el gráfico de barras con Seaborn usando la paleta personalizada
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x=conteo_por_tipo_via_sorted.values, y=conteo_por_tipo_via_sorted.index, palette=custom_palette)

# Añadir título y etiquetas
plt.title('Número de incidentes por tipo de vía')
plt.xlabel('Número de incidentes')
plt.ylabel('Tipo de vía')

# Añadir etiquetas con el número de víctimas en cada barra
for index, value in enumerate(conteo_por_tipo_via_sorted.values):
    barplot.text(value, index, str(value), ha='left', va='center', color='black')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Convertir todos los valores de la columna "momento_hecho" a minúsculas
df['momento_hecho'] = df['momento_hecho'].str.lower()

# Obtener los valores únicos en la columna "momento_hecho"
momentos_unicos = df['momento_hecho'].unique()

# Contar el número de incidentes por momento del día
conteo_por_momento = df['momento_hecho'].value_counts()

# Ordenar el conteo por el orden de los momentos del día
conteo_por_momento_sorted = conteo_por_momento.reindex(momentos_unicos)

# Crear una paleta de colores personalizada
custom_palette = sns.color_palette("husl", len(momentos_unicos))

# Crear el gráfico de barras con Seaborn usando la paleta personalizada
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x=conteo_por_momento_sorted.index, y=conteo_por_momento_sorted.values, hue=conteo_por_momento_sorted.index, palette=custom_palette, legend=False)

# Añadir título y etiquetas
plt.title('Número de incidentes por momento del día')
plt.xlabel('Momento del día')
plt.ylabel('Número de incidentes')

# Añadir etiquetas con el número de incidentes en cada barra
for index, value in enumerate(conteo_por_momento_sorted.values):
    barplot.text(index, value, str(value), ha='center', va='bottom', color='black')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(x='edad', y='sexo', data=df)
plt.title('Gráfico de Dispersión: Edad vs. Sexo')
plt.xlabel('Edad')
plt.ylabel('Sexo')
plt.show()

# Boxplot
sns.boxplot(x='sexo', y='edad', data=df)
plt.title('Boxplot: Edad por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Edad')
plt.show()

# Violin plot
sns.violinplot(x='sexo', y='edad', data=df)
plt.title('Violin Plot: Edad por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Edad')
plt.show()
```
